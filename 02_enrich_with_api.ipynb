```python
import requests 
import json
import pandas as pd
```

```python
url="https://open.er-api.com/v6/latest/EUR"
```

```python
# 1) Appel de l'API
response = requests.get(url)

# 2) Conversion de la reponse en dictionnaire Python
data = response.json()

# 3) Recuperer tous les taux
rates = data["rates"]

# 4) Afficher seulement les taux qui nous interessent
print("USD :", rates["USD"])
print("GBP :", rates["GBP"])
print("XOF :", rates["XOF"])
```

```python
clients = pd.read_csv('data/raw/client.csv')
clients
```

```python
COUNTRY_CURRENCY = {
    "France": "EUR",
    "Allemagne": "EUR",
    "Canada": "CAD",
    "Etats-Unis": "USD",
    "Bresil": "BRL",
    "Argentine": "ARS",
    "Maroc": "MAD",
    "Egypte": "EGP",
    "Nigeria": "NGN",
    "Afrique du Sud": "ZAR",
    "Kenya": "KES",
    "Japon": "JPY",
    "Chine": "CNY",
    "Inde": "INR",
    "Coree du Sud": "KRW",
    "Indonesie": "IDR",
    "Australie": "AUD",
    "Nouvelle-Zelande": "NZD",
    "Mexique": "MXN",
    "Turquie": "TRY"
}

# Version simple : on prend la devise directement depuis le pays
clients['Devise'] = clients['pays'].map(COUNTRY_CURRENCY)

# Si le pays n'est pas dans le dictionnaire, on met UNK
clients['Devise'] = clients['Devise'].fillna('UNK')

clients[['city', 'pays', 'Devise']]
```

```python
commandes = pd.read_csv('data/raw/commande.csv')
commandes
```

```python
usd_rate = rates.get('USD', 1.0)
commandes['total_usd'] = commandes['total_eur'] * usd_rate
commandes
```

```python
# 1) Nettoyer les colonnes pour eviter les erreurs si on relance la cellule
commandes = commandes.drop(columns=['Devise', 'fx_rate_local', 'total_local', 'id'], errors='ignore')

# 2) Prendre seulement id + devise dans clients
table_devise = clients[['id', 'Devise']]

# 3) Ajouter la devise dans commandes (grace au client_id)
commandes = commandes.merge(table_devise, left_on='client_id', right_on='id', how='left')

# 4) Si devise manquante, mettre UNK
commandes['Devise'] = commandes['Devise'].fillna('UNK')

# 5) Supprimer la colonne id ajoutee par le merge
commandes = commandes.drop(columns=['id'])

# 6) Preparer les taux
rates_local = rates.copy()
rates_local['EUR'] = 1.0

# 7) Ajouter le taux selon la devise
commandes['fx_rate_local'] = commandes['Devise'].map(rates_local).fillna(1.0)

# 8) Calculer le montant local
commandes['total_local'] = commandes['total_eur'] * commandes['fx_rate_local']

# 9) Afficher un apercu
commandes[['client_id', 'Devise', 'total_eur', 'fx_rate_local', 'total_local']].head()
```
